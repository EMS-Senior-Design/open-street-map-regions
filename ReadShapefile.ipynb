```python
find_county = re.compile("Census Tract (.*), Travis County, Texas")
```

```python
d = {}
```

```python
for s in shape:
    if s["properties"]["COUNTY"] == "Travis":
#         d[find_county.match(s["properties"]["LOCATION"]).group(1)] = s["geometry"]["coordinates"]
        d[find_county.match(s["properties"]["LOCATION"]).group(1).replace(".", "_")] = s["properties"]["FIPS"]
```

```python
len(d)
```

```python
for census_tract in os.listdir("census_tracts_2018"):
    with open("census_tracts_2018_fips/{}".format(d[census_tract[:-4]]+".txt"), "w") as w:
        with open("census_tracts_2018/{}".format(census_tract), "r") as r:
            w.write(r.read())
```

```python
url = "https://overpass-api.de/api/interpreter"
# l = os.listdir("census_tracts")
l = os.listdir("census_tracts_2018")
for key in d.keys():
    s = " ".join(map(lambda x: str(x[1]) + " " + str(x[0]), d[key][0]))
    fname = key.replace(".", "_")
    if (fname+".txt") in l:
        print("Already did {}".format(fname))
        continue
    print(fname)
    data = {"data": 'node(poly:\"{}\"); out;'.format(s)}
    resp = requests.post(url, data=data)
#     with open("census_tracts/{}.txt".format(fname), "w")as w:
    with open("census_tracts_2018/{}.txt".format(fname), "w")as w:
        w.write(resp.text)
```
