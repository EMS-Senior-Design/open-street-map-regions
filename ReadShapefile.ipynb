```python
url = "https://overpass-api.de/api/interpreter"
dir_name = "census_tracts_2018_fips"
# l = os.listdir("census_tracts")
l = os.listdir(dir_name)
for key in d.keys():
    s = " ".join(map(lambda x: str(x[1]) + " " + str(x[0]), d[key][0]))
    fname = key.replace(".", "_")
    if (fname+".txt") in l:
        print("Already did {}".format(fname))
        continue
    print(fname)
    data = {"data": 'node(poly:\"{}\"); out;'.format(s)}
    resp = requests.post(url, data=data)
    with open("{}/{}.txt".format(dir_name, fname), "w")as w:
        w.write(resp.text)
    time.sleep(5)
```
