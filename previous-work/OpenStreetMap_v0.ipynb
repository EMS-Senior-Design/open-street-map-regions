## OSM Code

```python
#pip install osmapi
```

```python
import osmapi
import pprint #for pretty printing of dict

api = osmapi.OsmApi()
```

```python
# Example
print(api.NodeGet(123))
# {u'changeset': 532907, u'uid': 14298,
# u'timestamp': u'2007-09-29T09:19:17Z',
# u'lon': 10.790009299999999, u'visible': True,
# u'version': 1, u'user': u'Mede',
# u'lat': 59.9503044, u'tag': {}, u'id': 123}
```

```python
nodeMap = api.Map(97, 30, 97.25, 30.25)
pprint.pprint(nodeMap)
```
